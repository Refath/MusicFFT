```python
import numpy as np
import matplotlib.pyplot as plt

def Harmonic(i,linComb):
    x=np.linspace(0,2*np.pi,2000)
    y = [0 for _ in x] 
    weights = linComb
    for n in range(0,i):
        y += weights[n]*np.sin((2*n+1)*(2*np.pi)*(x))/(2*n+1)
    plt.plot(x,y)
    plt.grid()
    plt.show()

# Harmonic (# HARMONICS, WEIGHTS)
# TRY CHANGING THE NUMBER OF HARMONICS AND THE WEIGHTS!
Harmonic(5,[5,3,2,0,0])
```
